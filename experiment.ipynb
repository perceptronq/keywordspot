```python
import torchaudio
import torch
import os

DATA_PATH = '/home/mayankch283/Downloads/speech_commands_v0.02'
    
sample_file = os.path.join(DATA_PATH, 'bed', '01648c51_nohash_1.wav')
print(sample_file)
waveform, sample_rate = torchaudio.load(sample_file)

print("Waveform Shape:", waveform.shape)
print("Sample Rate:", sample_rate)
```

```python
import torchaudio
import torchaudio.transforms as T
import matplotlib.pyplot as plt

# Convert waveform to mel spectrogram
mel_transform = T.MelSpectrogram(sample_rate=sample_rate, n_mels=64)
mel_spec = mel_transform(waveform)
log_mel_spec = torchaudio.transforms.AmplitudeToDB()(mel_spec)

# Visualize the mel spectrogram
plt.figure(figsize=(10, 4))
plt.imshow(log_mel_spec[0].detach().numpy(), aspect='auto', origin='lower', 
           extent=[0, log_mel_spec.size(1) / sample_rate, 0, 64])
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Mel Frequency')
plt.show()
```
